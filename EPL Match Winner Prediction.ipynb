```python
matches.dtypes  

## ML algorithms can only work with numeric data for example float64, int64 data types here. We need to figure out a way to covert other datatypes into numeric data.
```

```python
matches["date"] = pd.to_datetime(matches["date"])  #coverting date into datetime
matches
matches.dtypes
```

```python
matches["venue_code"] = matches["venue"].astype("category").cat.codes
matches  # 0 - Away   & 1- Home
```

```python
matches["opp_code"] = matches["opponent"].astype("category").cat.codes
matches
```

```python
matches["hour"] = matches["time"].str.replace(":.+", "", regex=True).astype("int")
matches
```

```python
matches["day_code"] = matches["date"].dt.dayofweek
matches
```

```python
matches["target"] = (matches["result"] == "W").astype("int")
matches
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
train = matches[matches["date"] < '2022-01-01']
test = matches[matches["date"] > '2022-01-01']
```

```python
test
```

```python
predictors = ["venue_code", "opp_code", "hour", "day_code"]
```

```python
rf.fit(train[predictors], train["target"])
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(test["target"], preds)
```

```python
acc
```

```python
combined = pd.DataFrame(dict(actual = test["target"], predictions = preds))
combined
```

```python
pd.crosstab(index = combined["actual"], columns = combined["predictions"])
```

we can see that the loss/draw was predicted accurately 141 times
An actual win was predicted as loss/draw 76 times
A loss/draw was predicted as win 31 times
A win was predicted as a win only 28 times

```python
from sklearn.metrics import precision_score
```

```python
precision_score(test["target"], preds)
```

precision of predictions is only 47.45%

```python
grouped_matches = matches.groupby("team")
```

```python
group = grouped_matches.get_group("Manchester City")
group
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stat = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stat
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_averages(group, cols, new_cols).head()
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team')
matches_rolling
```

It can be observed that there are no unique indices in the above dataframe (the first column with values 6,7,9 etc). We have 1320 rows. We need 1317 unique indices for those rows.

```python
matches_rolling.index = range(matches_rolling.shape[0])
matches_rolling
```

Retraining our model 

```python
train = matches_rolling[matches_rolling["date"] < '2022-01-01']
test = matches_rolling[matches_rolling["date"] > '2022-01-01']
```

```python
rf.fit(train[predictors + new_cols], train["target"])
```

```python
preds = rf.predict(test[predictors + new_cols])
```

```python
combined = pd.DataFrame(dict(actual = test["target"], predictions = preds))
```

```python
precision_score(test["target"], preds)
```

The precision score improved from 47% to 62% after considering rolling averages
